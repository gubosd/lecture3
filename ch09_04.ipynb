## Pivot Tables

```python
%pylab inline

import numpy as np
import pandas as pd
```

```python
tips = pd.read_csv('tips.csv')
tips['tip_pct'] = tips.tip / tips.total_bill
tips.head()
```

```python
tips.pivot_table(index=['sex','smoker']) # default is mean()
```

```python
tips.pivot_table('tip_pct',index=['sex','smoker'],aggfunc='mean')
#tips.groupby(['sex','smoker'])[['tip_pct']].mean()

tips.pivot_table('tip_pct',index='sex',columns='smoker',aggfunc='mean')
# tips.groupby(['sex','smoker']).tip_pct.mean().unstack()
```

```python
tips.pivot_table(['size','tip_pct'],index=['sex','day'],columns='smoker',margins=True)
```
