## Data Aggregation

```python
%pylab inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
df = pd.DataFrame({'data1': np.random.randn(5), 'data2': np.random.randn(5),
                   'key1': ['a','a','b','b','a'], 'key2': ['one','two','one','two','one']})
df
```

```python
df.groupby('key1').agg(lambda ser: ser.max()-ser.min())
```

```python
df.groupby('key1').describe()
```

## groupby methods
- count, size, sum, mean, median
- std, var, min, max
- prod, first, last

```python
tips = pd.read_csv('tips.csv')
display(tips.shape, tips.columns) # tips.dtypes
tips['tip_pct'] = tips.tip / tips.total_bill
tips.head()
```

```python
tips.groupby(['sex','smoker']).tip_pct.mean()
```

```python
tips.groupby(['sex','smoker']).tip_pct.agg(['mean','std','count',('gap',lambda s: s.max()-s.min())])
```

```python
tips.groupby(['sex','smoker']).agg(['mean','std'])
```
