## Resampling and Frequency Conversion

```python
%pylab inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
rng = pd.date_range('1/1/2018', periods=100, freq='D')
ts = pd.DataFrame(np.random.randint(10,size=[100,3]), index=rng, columns=list('abc'))
display(ts.tail())
ts.resample('MS').sum()
```

```python
ts2 = ts.set_index(pd.date_range('1/1/2018',periods=100,freq='T')) # 'T' = 'min'
display(ts2.head(10))
ts2.resample('H').sum()
```

```python
ts2.resample('3min').sum().head()
```

```python
ts2.resample('3min').ohlc().head()
```

```python
ts2.groupby(lambda t: t.hour).sum()
```
